```python
def inferred_causality(data):

  # Identify the Markov blankets for each variable in the data set.
  markov_blankets = {}
  for variable in data.columns:
    markov_blankets[variable] = identify_markov_blanket(data, variable)

  # Construct a causal graph from the Markov blankets.
  causal_graph = construct_causal_graph(markov_blankets)

  # Return the causal graph.
  return causal_graph

def identify_markov_blanket(data, variable):

  # Initialize the Markov blanket.
  markov_blanket = set()

  # Add all of the parents of the variable to the Markov blanket.
  for parent in data[variable].dropna().index:
    markov_blanket.add(parent)

  # Add all of the children of the variable to the Markov blanket.
  for child in data[variable].dropna().columns:
    markov_blanket.add(child)

  # Add all of the variables that are adjacent to the variable to the Markov blanket.
  for adjacent_variable in data.columns:
    if adjacent_variable != variable and adjacent_variable in markov_blankets[variable]:
      markov_blanket.add(adjacent_variable)

  # Return the Markov blanket.
  return markov_blanket

def construct_causal_graph(markov_blankets):

  # Initialize the causal graph.
  causal_graph = {}

  # For each variable in the data set, add an edge from each parent to the variable to the causal graph.
  for variable in markov_blankets.keys():
    for parent in markov_blankets[variable]:
      if parent not in causal_graph:
        causal_graph[parent] = set()
      causal_graph[parent].add(variable)

  # For each variable in the data set, add an edge from each child to the variable to the causal graph.
  for variable in markov_blankets.keys():
    for child in markov_blankets[variable]:
      if child not in causal_graph:
        causal_graph[child] = set()
      causal_graph[child].add(variable)

  # Return the causal graph.
  return causal_graph
```

```python
import yfinance as yf
import pandas as pd
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb
import matplotlib.pyplot as plt

class CrudeOilForecaster:
    def __init__(self, data, target):
        self.data = data.copy()
        self.target = target

        # Ensuring all columns are discretized and set as categorical
        for feature in self.data.columns:
            # Handling NaN values
            self.data[feature].fillna(method='ffill', inplace=True)

            # Check the number of unique values
            unique_vals = self.data[feature].nunique()

            # If there are at least 3 unique values, use quantile cut, otherwise use unique values as bins
            if unique_vals >= 3:
                self.data[feature] = pd.qcut(self.data[feature], 3, labels=['low', 'medium', 'high'], duplicates='drop')
            else:
                self.data[feature] = pd.cut(self.data[feature], bins=unique_vals, labels=range(unique_vals), duplicates='drop')

            # Ensuring the type is category
            self.data[feature] = self.data[feature].astype('category')

        # All other columns are considered features except target
        self.features = [col for col in self.data.columns if col != target]

        # Utilizing Hill Climbing for structure learning
        self.learner = gum.BNLearner(self.data)
        self.learner.useLocalSearchWithTabuList()

        # Get the Bayesian Network from the learner
        self.bn = self.learner.learnBN()

        # Learn the parameters (CPTs) of the BN
        self.learner.learnParameters(self.bn)

    def predict(self, new_features):
        # Inference with the learned Bayesian Network
        ie = gum.LazyPropagation(self.bn)
        ie.setEvidence(new_features)
        ie.makeInference()

        # Probability distribution of the target given new features
        prob_dist = ie.posterior(self.target)

        return prob_dist

    def plot_graph(self):
        # Utilize pyAgrum's plotting capabilities
        gnb.showBN(self.bn, size="20")
        plt.show()

if __name__ == '__main__':
    # Download historical market data for Crude Oil prices from Yahoo Finance
    natural_gas = yf.Ticker("NG=F")
    data = natural_gas.history(period="5y")

    # Define your target variable
    target = 'Close'

    # Initialize the forecaster
    forecaster = CrudeOilForecaster(data, target)

    # Test predictions
    # Note: Make sure that new_features contain the right labels, either [low, medium, high] or [0, 1], depending on your discretization
    new_features = {'Open': 'medium', 'High': 'medium', 'Low': 'low', 'Volume': 'high'}
    prediction = forecaster.predict(new_features)

    # Display the prediction
    print("Prediction: ", prediction)

    # Visualize the Bayesian Network
    forecaster.plot_graph()

```
